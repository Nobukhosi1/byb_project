# Data visualisation task 

**Additional information**  
[Cars93 metadata](https://jse.amstat.org/v1n1/datasets.lock.html) 

```python
# Import libraries
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Load data
car_data = pd.read_csv('Cars93.csv', index_col = 0, keep_default_na=False) 
```

### 1. Boxplot

```python
# Boxplot for revs per mile

#selected manufacturers
selected_manufacturers = ["Audi", "Hyndai", "Suzuki", "Toyota"]
filtered_df = car_data[car_data ["Manufacturer"].isin(selected_manufacturers)]    
# Select features for boxplot

plt.boxplot([filtered_df[filtered_df["Manufacturer"] == manuf] ["Rev.per.mile"] for manuf in selected_manufacturers], labels = selected_manufacturers)
plt.title("Rev per mile for selected car models")
plt.xlabel("Manufacturer")
plt.ylabel("Rev per mile")
```

**Boxplot question:**  
Which of these manufacturers has the car with the highest revs per mile?


**Boxplot answer:**
Toyata

### 2. Histogram

```python
# Histogram of MPG
plt.figure(figsize = (9, 5))

#Histogram of MPG in the city
plt.hist(car_data["MPG.city"], bins=20, alpha=0.5, color = "green", label= "MPG in city")

#Histogram of MPG on the highway
plt.hist(car_data["MPG.highway"], bins=20, alpha=0.5, color = "blue", label= "MPG in highway")

plt.title("MPG in the city and on the highway ")
plt.xlabel("MPG")
plt.ylabel("Frequency")
plt.legend()

```

**Histogram question:**  
Is it generally more fuel efficient to drive in the city or on the highway?

**Histogram answer:**
It is more fuel efficient to drive on the highway than in the city . This is shown by highway having more higher values towards higher miles per gallon(MPG) values compared to MPG in the city.

### 3. Lineplot

```python
# Lineplot (Wheelbase vs turning circle)
plt.figure(figsize = (10, 6))
plt.plot(car_data["Wheelbase"], car_data["Turn.circle"], marker="o", linestyle = "-")
plt.title("Relationship between Wheelbase and Turning Circle")
plt.xlabel("Wheelbase")
plt.ylabel("Turning Circle")
plt.grid(True)
```

**Lineplot question:**  
What is this relationship? What happens when the wheelbase gets larger?

**Lineplot answer:**  
It is a non linear relationship. As the whelbase gets larger, there are times that the turning cylcle also increases but there are also times where it decreases.

### 4. Barplot

```python
# Barplot for mean horsepower

mean_horsepower_by_type = car_data.groupby("Type")["Horsepower"].mean()

mean_horsepower_by_type.plot(kind="bar", color="red")
plt.title("Mean horsepower by car type")
plt.xlabel("Car type")
plt.ylabel("Mean horsepoer")
```

**Barplot question:**  
Does a larger car mean more horsepower?



**Barplot answer:**
Yes, this is because  large car type has the largest mean horsepower.
